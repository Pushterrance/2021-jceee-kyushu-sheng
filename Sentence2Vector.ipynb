## Library導入

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

## Annotation導入

```python
import json
import pandas as pd
json_file = './CTC_anns.json'
text_anno = json.load(open(json_file, 'r'))
text_anno.pop('dataset')
```

## 要らない部分を削除

```python
for item in text_anno['images']:
    del item['sentids']
    del item['partition']
    del item['imgid']
    del item['split']
    del item['coco_url']
    del item['gt_text']
for i in range(10683):
    for j in range(5):
        del text_anno['images'][i]['sentences'][j]['tokens']
        del text_anno['images'][i]['sentences'][j]['imgid']
        del text_anno['images'][i]['sentences'][j]['sentid']    
```

```python
tmp = json.dumps(text_anno)
tmp_function = open('sentence_CTC.json', 'w')
tmp_function.write(tmp)
jsonPath = './sentence_CTC.json'
with open(jsonPath,'r') as f: data=json.load(f)
```

## Sentence to Vector

```python
sentences = ['A suitcase that is next to a bag.',
    'A group of people riding boats along a river.', 
    'A double decker bus driving past a bunch of tall buildings.']
sentence_embeddings = model.encode(sentences)

for sentence, embedding in zip(sentences, sentence_embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
```

```python
sentences_suitcase = ['A suitcase that is next to a bag.',
    'A piece of luggage covered in luggage tags',
    'The luggage contains tags from several different locations.',
    'A piece of luggage that has paper tags on it.',
    'A suitcase is waiting to be claimed in an airport.']
sentence_embeddings = model.encode(sentences_suitcase)
suitcase_embeddings = []
for sentence, embedding in zip(sentences_suitcase, sentence_embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
    suitcase_embeddings.append(embedding)
```
